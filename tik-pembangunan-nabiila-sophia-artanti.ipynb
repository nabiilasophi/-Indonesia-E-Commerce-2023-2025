```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('/kaggle/input/datasets/bakitacos/indonesia-e-commerce-sales-and-shipping-20232025/all_months_clean.csv', sep=';')

print("Dimensi data kamu:", df.shape)
df.head()
```

```python
print("--- Informasi Struktur Data ---")
df.info()

print("\n--- Jumlah Data yang Kosong per Kolom ---")
print(df.isnull().sum())
```

```python
df['Waktu Pesanan Dibuat'] = pd.to_datetime(df['Waktu Pesanan Dibuat'], errors='coerce')

df['Alasan Pembatalan'] = df['Alasan Pembatalan'].fillna('Tidak Dibatalkan')

df['Tahun'] = df['Waktu Pesanan Dibuat'].dt.year
df['Bulan_Tahun'] = df['Waktu Pesanan Dibuat'].dt.to_period('M')

print("Pembersihan data selesai! Kolom baru 'Tahun' dan 'Bulan_Tahun' berhasil dibuat.")
```

```python
payment_counts = df['Metode Pembayaran'].value_counts()

plt.figure(figsize=(10, 5))
sns.barplot(x=payment_counts.values, y=payment_counts.index, palette='viridis')

plt.title('Distribusi Metode Pembayaran yang Digunakan Konsumen', fontsize=14, fontweight='bold')
plt.xlabel('Jumlah Transaksi', fontsize=12)
plt.ylabel('Metode Pembayaran', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
top_provinces = df['Provinsi'].value_counts().head(10)

plt.figure(figsize=(12, 5))
sns.barplot(x=top_provinces.index, y=top_provinces.values, color='#3266ad')

plt.title('Top 10 Provinsi dengan Transaksi E-Commerce Tertinggi', fontsize=14, fontweight='bold')
plt.xlabel('Provinsi', fontsize=12)
plt.ylabel('Jumlah Transaksi', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
monthly_trend = df.groupby('Bulan_Tahun')['Total Pembayaran'].sum().reset_index()
monthly_trend['Bulan_Tahun'] = monthly_trend['Bulan_Tahun'].astype(str)


plt.figure(figsize=(14, 5))
plt.plot(monthly_trend['Bulan_Tahun'], monthly_trend['Total Pembayaran'], marker='o', color='#2b5c8f', linewidth=2)

plt.title('Tren Perputaran Uang (Revenue) Bulanan E-Commerce', fontsize=14, fontweight='bold')
plt.xlabel('Periode (Bulan-Tahun)', fontsize=12)
plt.ylabel('Total Pembayaran (Rupiah)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
top_categories = df['product_categories'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_categories.values, y=top_categories.index, palette='magma')

plt.title('Top 10 Kategori Produk Paling Banyak Dipesan', fontsize=14, fontweight='bold')
plt.xlabel('Jumlah Pesanan', fontsize=12)
plt.ylabel('Kategori Produk', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
shipping_options = df['Opsi Pengiriman'].value_counts()

plt.figure(figsize=(8, 5))
plt.pie(shipping_options.values, labels=shipping_options.index, autopct='%1.1f%%', 
        colors=['#4f81bd', '#c0504d', '#9bbb59', '#8064a2'], startangle=140)

plt.title('Persentase Preferensi Opsi Pengiriman oleh Konsumen', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## Kesimpulan dan Analisis Perspektif TIK dalam Pembangunan

Berdasarkan Exploratory Data Analysis (EDA) yang telah dilakukan terhadap data transaksi E-Commerce Indonesia (2023–2025), berikut adalah kesimpulan strategis yang dikaitkan dengan aspek Teknologi Informasi dan Komunikasi (TIK) dalam Pembangunan:

### 1. Kesenjangan Digital (*Digital Divide*) dan Infrastruktur
* **Temuan Data:** Visualisasi top provinsi menunjukkan bahwa aktivitas transaksi e-commerce masih sangat didominasi oleh wilayah tertentu (khususnya Pulau Jawa/Jabodetabek) yang menjadi indikator nyata adanya *digital divide* atau kesenjangan digital di Indonesia. Infrastruktur internet yang andal dan penetrasi gawai yang belum merata di luar Pulau Jawa membuat adopsi ekonomi digital masih berpusat di wilayah urban. 

### 2. Inklusi Keuangan Digital (*Financial Inclusion*)
* **Temuan Data:** Tren penggunaan metode pembayaran non-tunai (seperti E-Wallet dan Transfer Bank) bersaing ketat dengan metode COD (*Cash on Delivery*). Tingginya penggunaan E-Wallet menunjukkan keberhasilan penetrasi *fintech* dalam meningkatkan inklusi keuangan pada masyarakat yang sebelumnya tidak memiliki akses ke bank (*unbanked population*). Namun, masih bertahannya metode COD menandakan bahwa tingkat kepercayaan digital sebagian masyarakat terhadap transaksi *online* masih perlu dibangun, atau akses layanan keuangan digital di daerah pelosok memang belum memadai.

### 3. Tantangan Logistik di Negara Kepulauan
* **Temuan Data:** Opsi pengiriman Reguler dan Hemat mendominasi preferensi konsumen dibandingkan opsi pengiriman cepat (*Same Day/Instant*). Konsumen cenderung memilih opsi yang lebih ekonomis demi menekan biaya ongkos kirim antarpulau yang tinggi. 

### 4. Pola Konsumerisme dan Pertumbuhan Ekonomi
* **Temuan Data:** Tren penjualan bulanan selalu mencapai puncak (*peak*) pada bulan November dan Desember. Kategori produk didominasi oleh kebutuhan sekunder/tersier.
